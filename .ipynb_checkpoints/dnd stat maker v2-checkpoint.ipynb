Dnd stat maker notes (2)
I think it’s becoming far too complicated. 
Let’s start from the top from how the PHB say to go.
1. Race
    * ASI
    * Age
    * Alignment
    * Size
    * Speed
    * Subrace
2. Class
    * Level
    * Hit points & Hit die
    * Proficiency bonus
3. Determine ability scores
    * This is when you roll
4. Background
5. Equipment

It seems like you record the ASI before rolling, so mods before rolling, and then assign and update it.

# Character class

```python
def gen_class(player_sheet):
    # , 'Wizard', 'Paladin', 'Rogue']
    samp_class_list = ['Barbarian', 'Fighter', 'Wizard']
    samp_class = random.choice(samp_class_list)
    # samp_class = 'Fighter'
    class_level = 1
    # class_hitpoint_base
    # class_hitdie
    class_proficiency_bonus = 2

    player_sheet['Proficiency bonus'] = class_proficiency_bonus
    player_sheet['Class'] = samp_class
    player_sheet['Level'] = 1
    class_hp_list = ['Hit Points', 'Hit Dice']
    player_sheet['Armor Class'] = 0
    if samp_class == 'Barbarian':
        class_hp_base = 12
        class_hitdie = '1d12 per barbarian level'
        player_sheet['Hit Points'] = dict(
            zip(class_hp_list, [class_hp_base, class_hitdie]))
        player_sheet['Other Proficiencies & Languages'][
            'Other proficiencies'].append('Light armor')
        player_sheet['Other Proficiencies & Languages'][
            'Other proficiencies'].append('Medium armor')
        player_sheet['Other Proficiencies & Languages'][
            'Other proficiencies'].append('Shields')
        player_sheet['Other Proficiencies & Languages'][
            'Other proficiencies'].append('Simple weapons')
        player_sheet['Other Proficiencies & Languages'][
            'Other proficiencies'].append('Martial weapons')
        #     player_sheet['Saving Throws'] = ['Strength', 'Constitution']
        player_sheet['Saving Throws']['Strength'][0] = player_sheet['Saving Throws']['Strength'][0] + \
            class_proficiency_bonus
        player_sheet['Saving Throws']['Strength'][1] = True
        player_sheet['Saving Throws']['Constitution'][0] = player_sheet['Saving Throws']['Constitution'][0] + \
            class_proficiency_bonus
        player_sheet['Saving Throws']['Constitution'][1] = True

        skill_list = [
            'Animal Handling', 'Athletics', 'Intimidation', 'Nature',
            'Perception', 'Survival'
        ]
        for itm in skill_list:
            if player_sheet['Skills'][itm][1] == True:
                skill_list.remove(itm)
        class_skills = random.sample(skill_list, k=2)
        for itm in class_skills:
            player_sheet['Skills'][itm][0] = player_sheet['Skills'][itm][0] + \
                class_proficiency_bonus
            player_sheet['Skills'][itm][1] = True

        player_sheet['Features & Traits'].append({
            'Rage': {
                '# Rages': 2,
                'Rage Damage': 2
            }
        })
        player_sheet['Features & Traits'].append('Unarmored Defense')

    elif samp_class == 'Fighter':
        class_hp_base = 10
        class_hitdie = '1d10 per fighter level'
        player_sheet['Hit Points'] = dict(
            zip(class_hp_list, [class_hp_base, class_hitdie]))
        player_sheet['Other Proficiencies & Languages'][
            'Other proficiencies'].append('All Armor')
        player_sheet['Other Proficiencies & Languages'][
            'Other proficiencies'].append('Shields')
        player_sheet['Other Proficiencies & Languages'][
            'Other proficiencies'].append('Simple weapons')
        player_sheet['Other Proficiencies & Languages'][
            'Other proficiencies'].append('Martial weapons')
        #     player_sheet['Saving Throws'] = ['Strength', 'Constitution']
        player_sheet['Saving Throws']['Strength'][0] = player_sheet['Saving Throws']['Strength'][0] + \
            class_proficiency_bonus
        player_sheet['Saving Throws']['Strength'][1] = True
        player_sheet['Saving Throws']['Constitution'][0] = player_sheet['Saving Throws']['Constitution'][0] + \
            class_proficiency_bonus
        player_sheet['Saving Throws']['Constitution'][1] = True

        skill_list = [
            'Acrobatics', 'Animal Handling', 'Athletics', 'History', 'Insight',
            'Intimidation', 'Perception', 'Survival'
        ]

        for itm in skill_list:
            if player_sheet['Skills'][itm][1] == True:
                skill_list.remove(itm)
        class_skills = random.sample(skill_list, k=2)
        for itm in class_skills:
            player_sheet['Skills'][itm][0] = player_sheet['Skills'][itm][0] + \
                class_proficiency_bonus
            player_sheet['Skills'][itm][1] = True
    elif samp_class == 'Wizard':
        class_hp_base = 6
        class_hitdie = '1d6 per wizard level'
        player_sheet['Hit Points'] = dict(
            zip(class_hp_list, [class_hp_base, class_hitdie]))
        player_sheet['Other Proficiencies & Languages'][
            'Other proficiencies'].append('Dagger')
        player_sheet['Other Proficiencies & Languages'][
            'Other proficiencies'].append('Dart')
        player_sheet['Other Proficiencies & Languages'][
            'Other proficiencies'].append('Sling')
        player_sheet['Other Proficiencies & Languages'][
            'Other proficiencies'].append('Quarterstaff')
        player_sheet['Other Proficiencies & Languages'][
            'Other proficiencies'].append('Light crossbow')
        player_sheet['Other Proficiencies & Languages'][
            'Other proficiencies'].append('Martial weapons')
        #     player_sheet['Saving Throws'] = ['Strength', 'Constitution']
        player_sheet['Saving Throws']['Intelligence'][0] = player_sheet['Saving Throws']['Intelligence'][0] + \
            class_proficiency_bonus
        player_sheet['Saving Throws']['Intelligence'][1] = True
        player_sheet['Saving Throws']['Wisdom'][0] = player_sheet['Saving Throws']['Wisdom'][0] + \
            class_proficiency_bonus
        player_sheet['Saving Throws']['Wisdom'][1] = True

        skill_list = [
            'Arcana', 'History', 'Insight',
            'Investigation', 'Medicine', 'Religion'
        ]

        for itm in skill_list:
            if player_sheet['Skills'][itm][1] == True:
                skill_list.remove(itm)
        class_skills = random.sample(skill_list, k=2)
        for itm in class_skills:
            player_sheet['Skills'][itm][0] = player_sheet['Skills'][itm][0] + \
                class_proficiency_bonus
            player_sheet['Skills'][itm][1] = True
        player_sheet['Spellcasting'] = {'Spellcasting Ability':{
            'Spell save DC':8 + class_proficiency_bonus,
            'Spell attack modifier': class_proficiency_bonus},
                                        'Cantrips': [3,'pick them'], 
                                        'Spellbook':{'1st level':6}, 
                                        'Prepared spells':1,
                                        'Spell slots':{'1st level':2}}
        player_sheet['Features & Traits'].append('Ritual Casting')
        player_sheet['Features & Traits'].append('Spellcasting focus: arcane')
        player_sheet['Features & Traits'].append('Arcane Recovery')
        
    else:
        print('not done')

    return  # player_sheet
```

# Determine ability scores

Roll time

```python
def roll():
    stat_list = []
    for i in range(0, 6):
        temp_stat_rolls = []
        for j in range(0,4):
            temp_stat_rolls.append(random.randint(1,6))
        temp_stat_rolls.remove(min(temp_stat_rolls))
        temp_stat = sum(temp_stat_rolls)
        stat_list.append(temp_stat)

    stat_list.sort()
#     print(mod_list, '\n', stat_list)
    return(stat_list)
```

```python
def gen_stats(player_sheet):
    roll_list = roll()
    print(roll_list)

    if player_sheet['Class'] == 'Barbarian':
        roll_str = roll_list.pop(roll_list.index(max(roll_list)))
        roll_con = roll_list.pop(roll_list.index(max(roll_list)))
        roll_dex = roll_list.pop(roll_list.index(max(roll_list)))
        roll_wis = roll_list.pop(roll_list.index(max(roll_list)))
        roll_cha = roll_list.pop(roll_list.index(max(roll_list)))
        roll_int = roll_list.pop(roll_list.index(max(roll_list)))


    elif player_sheet['Class'] == 'Fighter':
        roll_str = roll_list.pop(roll_list.index(max(roll_list)))
        roll_con = roll_list.pop(roll_list.index(max(roll_list)))
        roll_dex = roll_list.pop(roll_list.index(max(roll_list)))
        roll_int = roll_list.pop(roll_list.index(max(roll_list)))
        roll_cha = roll_list.pop(roll_list.index(max(roll_list)))
        roll_wis = roll_list.pop(roll_list.index(max(roll_list)))
        
    elif player_sheet['Class'] == 'Wizard':
        roll_int = roll_list.pop(roll_list.index(max(roll_list)))
        roll_con = roll_list.pop(roll_list.index(max(roll_list)))
        roll_dex = roll_list.pop(roll_list.index(max(roll_list)))
        roll_cha = roll_list.pop(roll_list.index(max(roll_list)))
        roll_wis = roll_list.pop(roll_list.index(max(roll_list)))
        roll_str = roll_list.pop(roll_list.index(max(roll_list)))
        

    player_sheet['Ability Scores']['Strength'][
        0] = player_sheet['Ability Scores']['Strength'][0] + roll_str
    player_sheet['Ability Scores']['Strength'][1] = (
        player_sheet['Ability Scores']['Strength'][0] // 2) - 5
    player_sheet['Ability Scores']['Dexterity'][
        0] = player_sheet['Ability Scores']['Dexterity'][0] + roll_dex
    player_sheet['Ability Scores']['Dexterity'][1] = (
        player_sheet['Ability Scores']['Dexterity'][0] // 2) - 5
    player_sheet['Ability Scores']['Constitution'][
        0] = player_sheet['Ability Scores']['Constitution'][0] + roll_con
    player_sheet['Ability Scores']['Constitution'][1] = (
        player_sheet['Ability Scores']['Constitution'][0] // 2) - 5
    player_sheet['Ability Scores']['Intelligence'][
        0] = player_sheet['Ability Scores']['Intelligence'][0] + roll_int
    player_sheet['Ability Scores']['Intelligence'][1] = (
        player_sheet['Ability Scores']['Intelligence'][0] // 2) - 5
    player_sheet['Ability Scores']['Wisdom'][
        0] = player_sheet['Ability Scores']['Wisdom'][0] + roll_wis
    player_sheet['Ability Scores']['Wisdom'][1] = (
        player_sheet['Ability Scores']['Wisdom'][0] // 2) - 5
    player_sheet['Ability Scores']['Charisma'][
        0] = player_sheet['Ability Scores']['Charisma'][0] + roll_cha
    player_sheet['Ability Scores']['Charisma'][1] = (
        player_sheet['Ability Scores']['Charisma'][0] // 2) - 5
    player_sheet['Hit Points']['Hit Points'] = player_sheet['Hit Points']['Hit Points'] + \
        player_sheet['Ability Scores']['Constitution'][1]
    player_sheet['Level'] = class_level
    
    if 'Unarmored Defense' in player_sheet['Features & Traits']:
        player_sheet['Armor Class'] = 10 + player_sheet['Ability Scores']['Dexterity'][1] + player_sheet['Ability Scores']['Constitution'][1]
    if 'Dwarven toughness' in player_sheet['Features & Traits']:
        player_sheet['Hit Points']['Hit Points'] = player_sheet['Hit Points']['Hit Points'] + 1
    mod_scores(player_sheet['Ability Scores'],
               player_sheet['Skills'], player_sheet['Saving Throws'])
    
    if player_sheet['Class'] == 'Wizard':
        player_sheet['Spellcasting']['Spellcasting Ability']['Spell save DC'] = player_sheet['Spellcasting']['Spellcasting Ability']['Spell save DC'] + player_sheet['Ability Scores']['Intelligence'][1]
        player_sheet['Spellcasting']['Spellcasting Ability']['Spell attack modifier'] = player_sheet['Spellcasting']['Spellcasting Ability']['Spell attack modifier'] + player_sheet['Ability Scores']['Intelligence'][1]
        player_sheet['Spellcasting']['Prepared spells'] = player_sheet['Spellcasting']['Prepared spells'] + player_sheet['Ability Scores']['Intelligence'][1]
    return# player_sheet
```

```python
def mod_scores(scores, skills, saves):
    '''
    scores, skills, saves, and hp are dicts. 
    Each corresponds to their respective values. 
    scores = Ability scores
    Skills = skills
    saves = Saving throws
    hp = Hit Points
    
    This applies the Ability Score Modifier to the skills. 
    '''
    str_mod = scores['Strength'][1]
    dex_mod = scores['Dexterity'][1]
    con_mod = scores['Constitution'][1]
    int_mod = scores['Intelligence'][1]
    wis_mod = scores['Wisdom'][1]
    cha_mod = scores['Charisma'][1]
    for skill in list(skills.keys()):
        if skills[skill][2] == 'Str':
            skills[skill][0] = skills[skill][0] + str_mod
        elif skills[skill][2] == 'Dex':
            skills[skill][0] = skills[skill][0] + dex_mod
        elif skills[skill][2] == 'Int':
            skills[skill][0] = skills[skill][0] + int_mod
        elif skills[skill][2] == 'Wis':
            skills[skill][0] = skills[skill][0] + wis_mod
        elif skills[skill][2] == 'Cha':
            skills[skill][0] = skills[skill][0] + cha_mod
        else:
            print(skill)
            
    for save in list(list(saves.keys())):
        if saves[save][2] == 'Str': 
            saves[save][0] = saves[save][0] + str_mod
        elif saves[save][2] == 'Dex': 
            saves[save][0] = saves[save][0] + dex_mod
        elif saves[save][2] == 'Con': 
            saves[save][0] = saves[save][0] + con_mod
        elif saves[save][2] == 'Int': 
            saves[save][0] = saves[save][0] + int_mod
        elif saves[save][2] == 'Wis': 
            saves[save][0] = saves[save][0] + wis_mod
        elif saves[save][2] == 'Cha': 
            saves[save][0] = saves[save][0] + cha_mod
        else:
            print(save)
            
        
    return
    
```
