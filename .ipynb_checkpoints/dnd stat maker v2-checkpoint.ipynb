```python
def dragonborn_ancestry(subrace, prof, con): #implement something for level
    if subrace in ['Black', 'Copper']:
        breath_damage = 'Acid'
        breath_resist = 'Acid'
        breath_range = '5 by 30 ft line'
        breath_save = 'Dexterity'
    elif subrace in ['Blue', 'Bronze']:
        breath_damage = 'Lightning'
        breath_resist = 'Lightning'
        breath_range = '5 by 30 ft line'
        breath_save = 'Dexterity'    
    
    elif subrace in ['Brass', 'Gold', 'Red']:
        breath_damage = 'Fire'
        breath_resist = 'Fire'
        breath_save = 'Dexterity'
        if subrace == 'Brass':
            breath_range = '5 by 30 ft line'
        else:
            breath_range = '15 ft cone'
            
    elif subrace in ['Green', 'Silver', 'White']:
        breath_save = 'Constitution'
        breath_range = '15 ft cone'
        if subrace == 'Green':
            breath_damage = 'Poison'
        else:
            breath_damage = 'Cold'
        breath_resist = breath_damage
            
    
    breath_wep = {'Breath weapon': {
        'Damage type': breath_damage,
        'Range': breath_range,
        'Saving throw': {
            'Stat': breath_save,
            'DC': 8 + prof + con  # prof + race_ASI['Constitution'][1]
        },
        'Damage': '2d6 on a failed save, half as much for a successful save' # this will change later for lvl
    }}
    return [{'Damage Resistance':breath_resist}, breath_wep]
```

```python
def pick_langs(sheet):
    n_choices = sheet['Other Proficiencies & Languages']['Languages'].count('pick one')
    if n_choices < 1:
        return sheet  # no languages to pick.
    else:
        while sheet['Other Proficiencies & Languages']['Languages'].count('pick one') > 0:
            sheet['Other Proficiencies & Languages']['Languages'].remove('pick one')
        possible_langs = [
            'Common', 'Dwarvish', 'Elvish', 'Giant', 'Gnomish', 'Goblin',
            'Halfling', 'Orc', 'Abyssal', 'Celestial', 'Draconic',
            'Deep Speech', 'Infernal', 'Primordial', 'Sylvan', 'Undercommon'
        ]
        pickable_langs = []
        for lang in possible_langs:
            if lang not in sheet['Other Proficiencies & Languages']['Languages']:
                pickable_langs.append(lang)

        sheet['Other Proficiencies & Languages']['Languages'].extend(
            random.sample(pickable_langs, k=n_choices))
        return sheet
```

```python
def roll():
    stat_list = []
    for i in range(0, 6):
        temp_stat_rolls = []
        for j in range(0, 4):
            temp_stat_rolls.append(random.randint(1, 6))
        temp_stat_rolls.remove(min(temp_stat_rolls))
        temp_stat = sum(temp_stat_rolls)
        stat_list.append(temp_stat)

    stat_list.sort()
    #     print(mod_list, '\n', stat_list)
    return (stat_list)
```

```python
def mod_scores(scores, skills, saves):
    '''
    scores, skills, saves, and hp are dicts. 
    Each corresponds to their respective values. 
    scores = Ability scores
    Skills = skills
    saves = Saving throws
    hp = Hit Points

    This applies the Ability Score Modifier to the skills. 
    '''
    str_mod = scores['Strength'][1]
    dex_mod = scores['Dexterity'][1]
    con_mod = scores['Constitution'][1]
    int_mod = scores['Intelligence'][1]
    wis_mod = scores['Wisdom'][1]
    cha_mod = scores['Charisma'][1]
    for skill in list(skills.keys()):
        if skills[skill][2] == 'Str':
            skills[skill][0] += str_mod
        elif skills[skill][2] == 'Dex':
            skills[skill][0] += dex_mod
        elif skills[skill][2] == 'Int':
            skills[skill][0] += int_mod
        elif skills[skill][2] == 'Wis':
            skills[skill][0] += wis_mod
        elif skills[skill][2] == 'Cha':
            skills[skill][0] += cha_mod
        else:
            print(skill)

    for save in list(list(saves.keys())):
        if saves[save][2] == 'Str':
            saves[save][0] += str_mod
        elif saves[save][2] == 'Dex':
            saves[save][0] += dex_mod
        elif saves[save][2] == 'Con':
            saves[save][0] += con_mod
        elif saves[save][2] == 'Int':
            saves[save][0] += int_mod
        elif saves[save][2] == 'Wis':
            saves[save][0] += wis_mod
        elif saves[save][2] == 'Cha':
            saves[save][0] += cha_mod
        else:
            print(save)

    return
```

```python
def gen_stats(sheet):
    roll_list = roll()

    if sheet['Class'] == 'Barbarian':
        roll_str = roll_list.pop(roll_list.index(max(roll_list)))
        roll_con = roll_list.pop(roll_list.index(max(roll_list)))
        roll_dex = roll_list.pop(roll_list.index(max(roll_list)))
        roll_wis = roll_list.pop(roll_list.index(max(roll_list)))
        roll_cha = roll_list.pop(roll_list.index(max(roll_list)))
        roll_int = roll_list.pop(roll_list.index(max(roll_list)))

    elif sheet['Class'] == 'Fighter':
        roll_str = roll_list.pop(roll_list.index(max(roll_list)))
        roll_con = roll_list.pop(roll_list.index(max(roll_list)))
        roll_dex = roll_list.pop(roll_list.index(max(roll_list)))
        roll_int = roll_list.pop(roll_list.index(max(roll_list)))
        roll_cha = roll_list.pop(roll_list.index(max(roll_list)))
        roll_wis = roll_list.pop(roll_list.index(max(roll_list)))

    elif sheet['Class'] == 'Ranger':
        roll_dex = roll_list.pop(roll_list.index(max(roll_list)))
        roll_wis = roll_list.pop(roll_list.index(max(roll_list)))
        roll_con = roll_list.pop(roll_list.index(max(roll_list)))
        roll_str = roll_list.pop(roll_list.index(max(roll_list)))
        roll_cha = roll_list.pop(roll_list.index(max(roll_list)))
        roll_int = roll_list.pop(roll_list.index(max(roll_list)))

    elif sheet['Class'] == 'Wizard':
        roll_int = roll_list.pop(roll_list.index(max(roll_list)))
        roll_con = roll_list.pop(roll_list.index(max(roll_list)))
        roll_dex = roll_list.pop(roll_list.index(max(roll_list)))
        roll_cha = roll_list.pop(roll_list.index(max(roll_list)))
        roll_wis = roll_list.pop(roll_list.index(max(roll_list)))
        roll_str = roll_list.pop(roll_list.index(max(roll_list)))

    else:
        print('not done', sheet)

    sheet['Ability Scores']['Strength'][0] += roll_str
    sheet['Ability Scores']['Strength'][1] = (
        sheet['Ability Scores']['Strength'][0] // 2) - 5
    sheet['Ability Scores']['Dexterity'][0] += roll_dex
    sheet['Ability Scores']['Dexterity'][1] = (
        sheet['Ability Scores']['Dexterity'][0] // 2) - 5
    sheet['Ability Scores']['Constitution'][0] += roll_con
    sheet['Ability Scores']['Constitution'][1] = (
        sheet['Ability Scores']['Constitution'][0] // 2) - 5
    sheet['Ability Scores']['Intelligence'][0] += roll_int
    sheet['Ability Scores']['Intelligence'][1] = (
        sheet['Ability Scores']['Intelligence'][0] // 2) - 5
    sheet['Ability Scores']['Wisdom'][0] += roll_wis
    sheet['Ability Scores']['Wisdom'][1] = (
        sheet['Ability Scores']['Wisdom'][0] // 2) - 5
    sheet['Ability Scores']['Charisma'][0] += roll_cha
    sheet['Ability Scores']['Charisma'][1] = (
        sheet['Ability Scores']['Charisma'][0] // 2) - 5

    sheet['Hit Points']['Max Hit Points'] += sheet['Ability Scores'][
        'Constitution'][1]

    if 'Unarmored Defense' in sheet['Features & Traits']:
        sheet['Armor Class'] = 10 + sheet['Ability Scores']['Dexterity'][
            1] + sheet['Ability Scores']['Constitution'][1]
    if 'Dwarven Toughness' in sheet['Features & Traits']:
        sheet['Hit Points']['Max Hit Points'] += 1 * sheet['Level']

    mod_scores(sheet['Ability Scores'], sheet['Skills'],
               sheet['Saving Throws'])

    if sheet['Class'] == 'Wizard':
        sheet['Spellcasting']['Spellcasting Ability'][
            'Spell save DC'] += sheet['Ability Scores']['Intelligence'][1]
        sheet['Spellcasting']['Spellcasting Ability'][
            'Spell attack modifier'] += sheet['Ability Scores'][
                'Intelligence'][1]
        sheet['Spellcasting']['Prepared spells'] += sheet['Ability Scores'][
            'Intelligence'][1]

    if 'Dragonborn' in re.findall('Dragonborn', sheet['Race']):
        for itm in sheet['Features & Traits']:
            if type(itm) == dict:
                x = sheet['Features & Traits'].index(itm)
                #                 print('Breath weapon' in sheet['Features & Traits'][x])
                if 'Breath weapon' in sheet['Features & Traits'][x]:
                    sheet['Features & Traits'][x]['Breath weapon'][
                        'Saving throw']['DC'] += sheet['Ability Scores'][
                            'Constitution'][1]

        # update breath weapons

    return sheet  # sheet
```

```python
def gen_equipment(sheet):
    profs = sheet['Other Proficiencies & Languages']['Other proficiencies']
    weaps = sheet['Equipment']['Weapons & Armor']['Weapons']
    if sheet['Class'] == 'Barbarian':
        weaps.append(random.choice(['greataxe', 'martial melee weapon']))
        if 'greataxe' in weaps:
            if 'greataxe' in profs or 'Martial weapons' in profs:
                bonus = sheet['Proficiency Bonus']
            else:
                bonus = 0
            sheet['Attacks & Spellcasting']['greataxe'] = {
                'Atk bonus':
                sheet['Ability Scores']['Strength'][1] + bonus,
                'Damage die':
                '1d12 slashing',
                'Dmg bonus':
                sheet['Ability Scores']['Strength'][1],
                'Properties': ['heavy', 'two-handed']
            }

        weaps.append(random.choice([{'handaxe': 2}, 'simple weapon']))
        weaps.append({'javelins': 4})
        sheet['Attacks & Spellcasting']['javelin'] = {
            'Atk bonus':
            sheet['Ability Scores']['Strength'][1] +
            sheet['Proficiency Bonus'],
            'Damage die':
            '1d6 piercing',
            'Dmg bonus':
            sheet['Ability Scores']['Strength'][1],
            'Properties': [{
                'thrown': 'range 30/120'
            }]
        }
        sheet['Equipment']['Tools'].append("Explorer's pack")

    elif sheet['Class'] == 'Fighter':
        fstyle = None
        for itm in sheet['Features & Traits']:
            if type(itm) == dict:
                x = sheet['Features & Traits'].index(itm)
                if 'Fighting Style' in sheet['Features & Traits'][x]:
                    fstyle = sheet['Features & Traits'][x]
                    break

        if sheet['Ability Scores']['Dexterity'][1] + 11 > 16 or fstyle[
                'Fighting Style'] == 'Archery':
            sheet['Equipment']['Weapons & Armor']['Armor'].append(
                'Leather armor')
            weaps.append('Longbow')
            sheet['Attacks & Spellcasting']['longbow'] = {
                'Atk bonus':
                sheet['Ability Scores']['Dexterity'][1] +
                sheet['Proficiency Bonus'],
                'Damage die':
                '1d8 piercing',
                'Dmg bonus':
                sheet['Ability Scores']['Dexterity'][1],
                'Properties': [
                    'ranged', 'heavy', 'two-handed',
                    {
                        'ammunition': 'range 150/600'
                    }
                ]
            }
            weaps.append({'arrows': 20})
            sheet['Armor Class'] += 11 + sheet['Ability Scores']['Dexterity'][1]
        else:
            sheet['Equipment']['Weapons & Armor']['Armor'].append(
                'Chain mail')
            sheet['Armor Class'] += 16

        if fstyle['Fighting Style'] == 'Dueling' or fstyle[
                'Fighting Style'] == 'Protection':
            weaps.append('Martial weapon')
            sheet['Equipment']['Weapons & Armor']['Armor'].append(
                'Shield')
            sheet['Armor Class'] += 2
        elif fstyle['Fighting Style'] == 'Two-Weapon Fighting':
            weaps.append('Martial weapon')
            weaps.append('Martial weapon')
        else:
            x = random.randint(0, 1)
            if x == 0:
                weaps.append('Martial weapon')
                sheet['Equipment']['Weapons & Armor']['Armor'].append(
                    'Shield')
                sheet['Armor Class'] += 2
            else:
                weaps.append('Martial weapon')
                weaps.append('Martial weapon')
        # should I have archery go for the light crossbow?
        weaps.append(random.choice(['light crossbow', {'handaxe': 2}]))
        for itm in weaps:
            if type(itm) == dict:
                if 'handaxe' in weaps[weaps.index(itm)]:
                    if 'handaxe' in profs or 'Simple weapons' in profs:
                        bonus = sheet['Proficiency Bonus']
                    else:
                        bonus = 0
                    sheet['Attacks & Spellcasting']['handaxe'] = {
                        'Atk bonus':
                        sheet['Ability Scores']['Strength'][1] + bonus,
                        'Damage die':
                        '1d6 slashing',
                        'Dmg bonus':
                        sheet['Ability Scores']['Strength'][1],
                        'Properties': ['light', {
                            'thrown': 'range 20/60'
                        }]
                    }
                    break

        if 'light crossbow' in weaps:
            if 'light crossbow' in profs or 'Simple weapons' in profs:
                bonus = sheet['Proficiency Bonus']
            else:
                bonus = 0

            sheet['Attacks & Spellcasting']['light crossbow'] = {
                'Atk bonus':
                sheet['Ability Scores']['Dexterity'][1] + bonus,
                'Damage die':
                '1d8 slashing',
                'Dmg bonus':
                sheet['Ability Scores']['Dexterity'][1],
                'Properties': [
                    'ranged', {'ammunition': 'range 80/320'}, 
                    'loading', 'two-handed'
                ]
            }
            weaps.append({'bolts': 20})

        sheet['Equipment']['Tools'].append(
            random.choice(["Dungeoneer's pack", "Explorer's pack"]))

    elif sheet['Class'] == 'Wizard':
        if sheet['Ability Scores']['Strength'][1] > sheet[
                'Ability Scores']['Dexterity'][1]:
            weaps.append('quarterstaff')

        else:
            weaps.append(random.choice(['quarterstaff', 'dagger']))
        if 'quarterstaff' in weaps:
            sheet['Attacks & Spellcasting']['quarterstaff'] = {
                'Atk bonus':
                sheet['Ability Scores']['Strength'][1] +
                sheet['Proficiency Bonus'],
                'Damage die':
                '1d6 bludgeoning',
                'Dmg bonus':
                sheet['Ability Scores']['Strength'][1],
                'Properties': [{
                    'versatile': '1d8'
                }]
            }
        elif 'dagger' in weaps:
            fin_s = sheet['Ability Scores']['Strength'][
                1] + sheet['Proficiency Bonus']
            fin_d = sheet['Ability Scores']['Dexterity'][
                1] + sheet['Proficiency Bonus']
            if fin_s > fin_d:
                fin = fin_s
            else:
                fin = fin_d

            sheet['Attacks & Spellcasting']['dagger'] = {
                'Atk bonus': fin,
                'Damage die': '1d4 piercing',
                'Dmg bonus': fin - sheet['Proficiency Bonus'],
                'Properties': ['finesse', 'light', {
                    'thrown': 'range 20/60'
                }]
            }

        wiz_focus = random.choice(
            ['component pouch', 'crystal', 'orb', 'rod', 'staff', 'wand'])
        sheet['Equipment']['Gear'].append(wiz_focus)
        sheet['Equipment']['Tools'].append(
            random.choice(["Scholar's pack", "Explorer's pack"]))
        sheet['Equipment']['Gear'].append('spellbook')

    return sheet
```
