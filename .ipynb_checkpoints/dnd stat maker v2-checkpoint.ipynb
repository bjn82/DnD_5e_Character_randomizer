Dnd stat maker notes (2)
I think it’s becoming far too complicated. 
Let’s start from the top from how the PHB say to go.
1. Race
    * ASI
    * Age
    * Alignment
    * Size
    * Speed
    * Subrace
2. Class
    * Level
    * Hit points & Hit die
    * Proficiency bonus
3. Determine ability scores
    * This is when you roll
4. Background
5. Equipment

It seems like you record the ASI before rolling, so mods before rolling, and then assign and update it.

```python
def gen_player():
    player_char = [
        'Name', 'Sex', 'Race', 'Ability Scores', 'Age', 'Alignment', 'Speed', 'Size',
        'Saving Throws', 'Skills', 'Features & Traits', 'Other Proficiencies & Languages'
    ]
    samp_race_list = ['Human', 'Elf', 'Dwarf', 'Halfling']
    race_ASI = {
        'Strength': [0,0],
        'Dexterity': [0,0],
        'Constitution': [0,0],
        'Intelligence': [0,0],
        'Wisdom': [0,0],
        'Charisma': [0,0]
    }
    class_proficiency_bonus = 2
    race_traits = []  # features and traits
    race_profs = []  # languages & other proficiencies
    # race_skills = [] # THis was the orginal. If it doesn't work, 
    race_skills = {
        'Acrobatics': [0, False, 'Dex'],
        'Animal Handling': [0,False, 'Wis'],
        'Arcana': [0,False,'Int'], 
        'Athletics': [0,False,'Str'],
        'Deception': [0, False, 'Cha'],
        'History': [0, False, 'Int'], 
        'Insight': [0, False, 'Wis'],
        'Intimidation': [0, False, 'Cha'], 
        'Investigation': [0, False, 'Int'], 
        'Medicine': [0, False, 'Wis'], 
        'Nature': [0, False, 'Int'], 
        'Perception': [0, False, 'Wis'], 
        'Performance': [0, False, 'Cha'], 
        'Persuasion': [0, False, 'Cha'], 
        'Religion': [0, False, 'Int'], 
        'Sleight of Hand': [0, False, 'Dex'], 
        'Stealth': [0, False, 'Dex'], 
        'Survival': [0, False, 'Wis']
    }
    race_saving_throws = {
        'Strength': [0, False, 'Str'], 
        'Dexterity': [0, False, 'Dex'], 
        'Constitution': [0, False, 'Con'], 
        'Intelligence': [0, False, 'Int'], 
        'Wisdom': [0, False, 'Wis'], 
        'Charisma': [0, False, 'Cha']
    }
    samp_race = random.choice(samp_race_list)
    if samp_race == 'Dwarf':
        sex = random.choice(['M', 'F', 'n/a'])
        name = dwarf_name(sex)
        race_height = random.randint(4, 5)
        race_speed = 25
        race_weight = random.randint(125, 175)
        langs = ['Common', 'Dwarvish']
        race_size = 'Medium'
        race_ASI['Constitution'][0] = 2
        race_alignment = 'Lawful Good'
        race_age = random.randint(40, 300)
        race_traits.append('Darkvision 60ft')
        race_traits.append('Dwarven Resilience')
        race_traits.append('Stonecutting')
        race_profs = race_profs + [
            'battleaxe', 'handaxe', 'throwing hammer', 'warhammer'
        ]
        race_profs.append('Not slowed by heavy armor')
        race_profs.append(
            random.choice(["smith's tools", "brewer's supplies", "mason's tools"]))
        sheet_profs = dict(
            zip(['Languages', 'Other proficiencies'], [langs, race_profs]))
        subrace = random.choice(['hill', 'mountain'])
        if subrace == 'hill':
            race_ASI['Wisdom'][0] = 1
            race_traits.append('Dwarven toughness')
        elif subrace == 'mountain':
            race_ASI['Strength'][0] = 2
            race_profs.append('Light armor')
            race_profs.append('Medium armor')
        else:
            print('no homebrew')

        player_sheet = dict(
            zip(player_char, [
                name, sex, samp_race + ' (' + subrace + ')', race_ASI, race_age,
                race_alignment, race_speed, race_size, race_saving_throws, race_skills, race_traits,
                dict(
                    zip(['Languages', 'Other proficiencies'], [langs, race_profs]))
            ]))
    elif samp_race == 'Elf':
        sex = random.choice(['M', 'F', 'n/a'])
        race_height = random.randint(4, 5)
        race_speed = 30
        race_weight = random.randint(125, 175)
        langs = ['Common', 'Elvish']
        race_size = 'Medium'
        race_ASI['Dexterity'][0] = 2
        race_age = random.randint(40, 300)
        name = elf_name(sex, race_age)
        race_traits.append('Keen Senses')
    #     race_skills.append('Perception')
        race_skills['Perception'][0] = race_skills['Perception'][0] + class_proficiency_bonus
        race_skills['Perception'][1] = True
        race_traits.append('Fey Ancestry')
        race_traits.append('Trance')
        sheet_profs = dict(
            zip(['Languages', 'Other proficiencies'], [langs, race_profs]))
        subrace = random.choice(['high', 'wood', 'dark'])
        if subrace == 'high':
            race_ASI['Intelligence'][0] = 1
            race_traits.append('Darkvision 60ft')
            race_profs = race_profs + ['longsword', 'shortsword', 'longbow', 'shortbow']
            race_traits.append('1 wizard cantrip (int)')
            langs.append('pick one')
            race_alignment = 'Chaotic good'
        elif subrace == 'wood':
            race_ASI['Wisdom'][0] = 1
            race_traits.append('Darkvision 60ft')
            race_profs = race_profs + ['longsword', 'shortsword', 'longbow', 'shortbow']
            race_traits.append('Fleet of Foot')
            race_speed = 35
            race_traits.append('Mask of the Wild')
            race_alignment = 'Chaotic good'
        elif subrace == 'dark':
            race_ASI['Charisma'][0] = 1
            race_traits.append('Darkvision 120ft')
            race_traits.append('Sunlight sensitivity') 
            # what about making this a dictionary, with the value as the property
            race_traits.append('Drow magic')
            race_profs = race_profs + ['rapier', 'shortsword', 'hand crossbows']
            race_alignment = 'Chaotic evil'        
        else:
            print('no homebrew')

        player_sheet = dict(
            zip(player_char, [
                name, sex, samp_race + ' (' + subrace + ')', race_ASI, race_age,
                race_alignment, race_speed, race_size, race_saving_throws, race_skills, race_traits,
                dict(
                    zip(['Languages', 'Other proficiencies'], [langs, race_profs]))
            ]))
    elif samp_race == 'Human':
        sex = random.choice(['M', 'F', 'n/a'])
        subrace = random.choice([
            'Calishite', 'Chondathan', 'Damaran', 'Illuskan', 'Mulan', 'Rashemi',
            'Shou', 'Tethyrian', 'Turami'
        ])
        name = human_name(sex, subrace)
        race_height = random.randint(4, 7)
        race_speed = 30
        race_weight = random.randint(85, 240)
        langs = ['Common', 'pick one']
        race_size = 'Medium'
        race_ASI['Constitution'][0] = 1
        race_ASI['Strength'][0] = 1
        race_ASI['Dexterity'][0] = 1
        race_ASI['Intelligence'][0] = 1
        race_ASI['Wisdom'][0] = 1
        race_ASI['Charisma'][0] = 1
        race_alignment = 'neutral'
        race_age = random.randint(16, 85)
        sheet_profs = dict(
            zip(['Languages', 'Other proficiencies'], [langs, race_profs]))
        player_sheet = dict(
            zip(player_char, [
                name, sex, samp_race + ' (' + subrace + ')', race_ASI, race_age,
                race_alignment, race_speed, race_size, race_saving_throws, race_skills, race_traits,
                dict(
                    zip(['Languages', 'Other proficiencies'], [langs, race_profs]))
            ]))
    elif samp_race == 'Halfling':
        sex = random.choice(['M', 'F', 'n/a'])
        name = halfling_name(sex)
        race_height = 3
        race_speed = 25
        race_weight = random.randint(35, 45)
        langs = ['Common', 'Halfling']
        race_size = 'Small'
        race_ASI['Dexterity'][0] = 2
        race_alignment = 'Lawful Good'
        race_age = random.randint(15, 275)
        race_traits.append('Lucky')
        race_traits.append('Halfling Nimbleness')
        race_traits.append('Brave')
        
        subrace = random.choice(['Lightfoot', 'Stout'])
        
        if subrace == 'Lightfoot':
            race_ASI['Charisma'][0] = 1
            race_traits.append('Naturally Stealthy')
        elif subrace == 'Stout':
            race_ASI['Constitution'][0] = 1
            race_traits.append('Stout Resilience')
            
        player_sheet = dict(
            zip(player_char, [
                name, sex, samp_race + ' (' + subrace + ')', race_ASI, race_age,
                race_alignment, race_speed, race_size, race_saving_throws, race_skills, race_traits,
                dict(
                    zip(['Languages', 'Other proficiencies'], [langs, race_profs]))
            ]))
        
    else:
        print('not')
    
    return player_sheet
```

```python
def gen_class(player_sheet):
    samp_class_list = ['Barbarian', 'Fighter']#, 'Wizard', 'Paladin', 'Rogue']
    samp_class = random.choice(samp_class_list)
    # samp_class = 'Fighter'
    class_level = 1
    # class_hitpoint_base
    # class_hitdie
    class_proficiency_bonus = 2

    player_sheet['Proficiency bonus'] = class_proficiency_bonus
    player_sheet['Class'] = samp_class
    player_sheet['Level'] = 1
    class_hp_list = ['Hit Points', 'Hit Dice']
    if samp_class == 'Barbarian':
        class_hp_base = 12
        class_hitdie = '1d12 per barbarian level'
        player_sheet['Hit Points'] = dict(
            zip(class_hp_list, [class_hp_base, class_hitdie]))
        player_sheet['Other Proficiencies & Languages'][
            'Other proficiencies'].append('Light armor')
        player_sheet['Other Proficiencies & Languages'][
            'Other proficiencies'].append('Medium armor')
        player_sheet['Other Proficiencies & Languages'][
            'Other proficiencies'].append('Shields')
        player_sheet['Other Proficiencies & Languages'][
            'Other proficiencies'].append('Simple weapons')
        player_sheet['Other Proficiencies & Languages'][
            'Other proficiencies'].append('Martial weapons')
        #     player_sheet['Saving Throws'] = ['Strength', 'Constitution']
        player_sheet['Saving Throws']['Strength'][0] = player_sheet['Saving Throws']['Strength'][0] + \
            class_proficiency_bonus
        player_sheet['Saving Throws']['Strength'][1] = True
        player_sheet['Saving Throws']['Constitution'][0] = player_sheet['Saving Throws']['Constitution'][0] + \
            class_proficiency_bonus
        player_sheet['Saving Throws']['Constitution'][1] = True

        skill_list = [
            'Animal Handling', 'Athletics',
            'Intimidation', 'Nature', 'Perception', 'Survival'
        ]
        for itm in skill_list:
            if player_sheet['Skills'][itm][1] == True:
                skill_list.remove(itm)
        class_skills = random.sample(skill_list, k=2)
        for itm in class_skills:
            player_sheet['Skills'][itm][0] = player_sheet['Skills'][itm][0] + class_proficiency_bonus
            player_sheet['Skills'][itm][1] = True

        player_sheet['Features & Traits'].append({'Rage': {'# Rages':2, 'Rage Damage': 2}})
        player_sheet['Features & Traits'].append('Unarmored Defense')

    elif samp_class == 'Fighter':
        class_hp_base = 10
        class_hitdie = '1d10 per fighter level'
        player_sheet['Hit Points'] = dict(
            zip(class_hp_list, [class_hp_base, class_hitdie]))
        player_sheet['Other Proficiencies & Languages'][
            'Other proficiencies'].append('All Armor')
        player_sheet['Other Proficiencies & Languages'][
            'Other proficiencies'].append('Shields')
        player_sheet['Other Proficiencies & Languages'][
            'Other proficiencies'].append('Simple weapons')
        player_sheet['Other Proficiencies & Languages'][
            'Other proficiencies'].append('Martial weapons')
        #     player_sheet['Saving Throws'] = ['Strength', 'Constitution']
        player_sheet['Saving Throws']['Strength'][0] = player_sheet['Saving Throws']['Strength'][0] + \
            class_proficiency_bonus
        player_sheet['Saving Throws']['Strength'][1] = True
        player_sheet['Saving Throws']['Constitution'][0] = player_sheet['Saving Throws']['Constitution'][0] + \
            class_proficiency_bonus
        player_sheet['Saving Throws']['Constitution'][1] = True

        skill_list = [
            'Acrobatics', 'Animal Handling', 'Athletics', 'History', 'Insight',
            'Intimidation', 'Perception', 'Survival'
        ]

        for itm in skill_list:
            if player_sheet['Skills'][itm][1] == True:
                skill_list.remove(itm)
        class_skills = random.sample(skill_list, k=2)
        for itm in class_skills:
            player_sheet['Skills'][itm][0] = player_sheet['Skills'][itm][0] + class_proficiency_bonus
            player_sheet['Skills'][itm][1] = True

    else:
        print('not done')

    return# player_sheet
```

# Determine ability scores

Roll time

```python
def roll():
    stat_list = []
    for i in range(0, 6):
        temp_stat_rolls = []
        for j in range(0,4):
            temp_stat_rolls.append(random.randint(1,6))
        temp_stat_rolls.remove(min(temp_stat_rolls))
        temp_stat = sum(temp_stat_rolls)
        stat_list.append(temp_stat)

    stat_list.sort()
#     print(mod_list, '\n', stat_list)
    return(stat_list)
```

```python
def gen_stats(player_sheet):
    roll_list = roll()
    print(roll_list)

    if player_sheet['Class'] == 'Barbarian':
        roll_str = roll_list.pop(roll_list.index(max(roll_list)))
        roll_con = roll_list.pop(roll_list.index(max(roll_list)))
        roll_dex = roll_list.pop(roll_list.index(max(roll_list)))
        roll_wis = roll_list.pop(roll_list.index(max(roll_list)))
        roll_cha = roll_list.pop(roll_list.index(max(roll_list)))
        roll_int = roll_list.pop(roll_list.index(max(roll_list)))


    elif player_sheet['Class'] == 'Fighter':
        roll_str = roll_list.pop(roll_list.index(max(roll_list)))
        roll_con = roll_list.pop(roll_list.index(max(roll_list)))
        roll_dex = roll_list.pop(roll_list.index(max(roll_list)))
        roll_int = roll_list.pop(roll_list.index(max(roll_list)))
        roll_cha = roll_list.pop(roll_list.index(max(roll_list)))
        roll_wis = roll_list.pop(roll_list.index(max(roll_list)))


    player_sheet['Ability Scores']['Strength'][
        0] = player_sheet['Ability Scores']['Strength'][0] + roll_str
    player_sheet['Ability Scores']['Strength'][1] = (
        player_sheet['Ability Scores']['Strength'][0] // 2) - 5
    player_sheet['Ability Scores']['Dexterity'][
        0] = player_sheet['Ability Scores']['Dexterity'][0] + roll_dex
    player_sheet['Ability Scores']['Dexterity'][1] = (
        player_sheet['Ability Scores']['Dexterity'][0] // 2) - 5
    player_sheet['Ability Scores']['Constitution'][
        0] = player_sheet['Ability Scores']['Constitution'][0] + roll_con
    player_sheet['Ability Scores']['Constitution'][1] = (
        player_sheet['Ability Scores']['Constitution'][0] // 2) - 5
    player_sheet['Ability Scores']['Intelligence'][
        0] = player_sheet['Ability Scores']['Intelligence'][0] + roll_int
    player_sheet['Ability Scores']['Intelligence'][1] = (
        player_sheet['Ability Scores']['Intelligence'][0] // 2) - 5
    player_sheet['Ability Scores']['Wisdom'][
        0] = player_sheet['Ability Scores']['Wisdom'][0] + roll_wis
    player_sheet['Ability Scores']['Wisdom'][1] = (
        player_sheet['Ability Scores']['Wisdom'][0] // 2) - 5
    player_sheet['Ability Scores']['Charisma'][
        0] = player_sheet['Ability Scores']['Charisma'][0] + roll_cha
    player_sheet['Ability Scores']['Charisma'][1] = (
        player_sheet['Ability Scores']['Charisma'][0] // 2) - 5
    player_sheet['Hit Points']['Hit Points'] = player_sheet['Hit Points']['Hit Points'] + \
        player_sheet['Ability Scores']['Constitution'][1]
    player_sheet['Level'] = class_level

    mod_scores(player_sheet['Ability Scores'],
               player_sheet['Skills'], player_sheet['Saving Throws'])

    return# player_sheet
```

```python
gen_stats(player_sheet)
```

```python
def mod_scores(scores, skills, saves):
    '''
    scores, skills, saves, and hp are dicts. 
    Each corresponds to their respective values. 
    scores = Ability scores
    Skills = skills
    saves = Saving throws
    hp = Hit Points
    
    This applies the Ability Score Modifier to the skills. 
    '''
    str_mod = scores['Strength'][1]
    dex_mod = scores['Dexterity'][1]
    con_mod = scores['Constitution'][1]
    int_mod = scores['Intelligence'][1]
    wis_mod = scores['Wisdom'][1]
    cha_mod = scores['Charisma'][1]
    for skill in list(skills.keys()):
        if skills[skill][2] == 'Str':
            skills[skill][0] = skills[skill][0] + str_mod
        elif skills[skill][2] == 'Dex':
            skills[skill][0] = skills[skill][0] + dex_mod
        elif skills[skill][2] == 'Int':
            skills[skill][0] = skills[skill][0] + int_mod
        elif skills[skill][2] == 'Wis':
            skills[skill][0] = skills[skill][0] + wis_mod
        elif skills[skill][2] == 'Cha':
            skills[skill][0] = skills[skill][0] + cha_mod
        else:
            print(skill)
            
    for save in list(list(saves.keys())):
        if saves[save][2] == 'Str': 
            saves[save][0] = saves[save][0] + str_mod
        elif saves[save][2] == 'Dex': 
            saves[save][0] = saves[save][0] + dex_mod
        elif saves[save][2] == 'Con': 
            saves[save][0] = saves[save][0] + con_mod
        elif saves[save][2] == 'Int': 
            saves[save][0] = saves[save][0] + int_mod
        elif saves[save][2] == 'Wis': 
            saves[save][0] = saves[save][0] + wis_mod
        elif saves[save][2] == 'Cha': 
            saves[save][0] = saves[save][0] + cha_mod
        else:
            print(save)
            
        
    return
    
```

```python
def personality_trait_gen(bg):
    bg_roll = random.randint(1,8)
    if bg == 'Noble':
        if bg_roll == 1:
            return 'My eloquent flattery makes everyone I talk to feel like the most wonderful and important person in the world.'
        elif bg_roll == 2:
            return 'The common folk love me for my kindness and generosity.'
        elif bg_roll == 3:
            return 'No one could doubt by looking at my regal bearing that I am a cut above the unwashed masses.'
        elif bg_roll == 4:
            return 'I take great pains to always look my best and follow the latest fashions.'
        elif bg_roll == 5: 
            return 'I don’t like to get my hands dirty, and I won’t be caught dead in unsuitable accommodations.'
        elif bg_roll == 6: 
            return 'Despite my noble birth, I do not place myself above other folk. We all have the same blood.'
        elif bg_roll == 7: 
            return 'My favor, once lost, is lost forever.'
        elif bg_roll == 8:
            return 'If you do me an injury, I will crush you, ruin your name, and salt your fields.'
        else:
            return 'something went wrong'
    else:
        return 'working on it'
```
