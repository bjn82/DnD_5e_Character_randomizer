I think it’s becoming far too complicated. 
Let’s start from the top from how the PHB say to go.
1. Race
    * ASI
    * Age
    * Alignment
    * Size
    * Speed
    * Subrace
2. Class
    * Level
    * Hit points & Hit die
    * Proficiency bonus
3. Determine ability scores
    * This is when you roll
4. Background
5. Equipment

It seems like you record the ASI before rolling, so mods before rolling, and then assign and update it.

```python
def create_char():
    player_sheet = gen_player()
    gen_class(player_sheet)
    gen_stats(player_sheet)
    gen_bg(player_sheet)
    gen_equipment(player_sheet)
    if player_sheet['Class'] in ['Fighter', 'Paladin', 'Ranger']:
        apply_fstyle(player_sheet)
    return player_sheet
```

```python
def roll():
    stat_list = []
    for i in range(0, 6):
        temp_stat_rolls = []
        for j in range(0,4):
            temp_stat_rolls.append(random.randint(1,6))
        temp_stat_rolls.remove(min(temp_stat_rolls))
        temp_stat = sum(temp_stat_rolls)
        stat_list.append(temp_stat)

    stat_list.sort()
#     print(mod_list, '\n', stat_list)
    return(stat_list)
```

```python
def mod_scores(scores, skills, saves):
    '''
    scores, skills, saves, and hp are dicts. 
    Each corresponds to their respective values. 
    scores = Ability scores
    Skills = skills
    saves = Saving throws
    hp = Hit Points
    
    This applies the Ability Score Modifier to the skills. 
    '''
    str_mod = scores['Strength'][1]
    dex_mod = scores['Dexterity'][1]
    con_mod = scores['Constitution'][1]
    int_mod = scores['Intelligence'][1]
    wis_mod = scores['Wisdom'][1]
    cha_mod = scores['Charisma'][1]
    for skill in list(skills.keys()):
        if skills[skill][2] == 'Str':
            skills[skill][0] = skills[skill][0] + str_mod
        elif skills[skill][2] == 'Dex':
            skills[skill][0] = skills[skill][0] + dex_mod
        elif skills[skill][2] == 'Int':
            skills[skill][0] = skills[skill][0] + int_mod
        elif skills[skill][2] == 'Wis':
            skills[skill][0] = skills[skill][0] + wis_mod
        elif skills[skill][2] == 'Cha':
            skills[skill][0] = skills[skill][0] + cha_mod
        else:
            print(skill)
            
    for save in list(list(saves.keys())):
        if saves[save][2] == 'Str': 
            saves[save][0] = saves[save][0] + str_mod
        elif saves[save][2] == 'Dex': 
            saves[save][0] = saves[save][0] + dex_mod
        elif saves[save][2] == 'Con': 
            saves[save][0] = saves[save][0] + con_mod
        elif saves[save][2] == 'Int': 
            saves[save][0] = saves[save][0] + int_mod
        elif saves[save][2] == 'Wis': 
            saves[save][0] = saves[save][0] + wis_mod
        elif saves[save][2] == 'Cha': 
            saves[save][0] = saves[save][0] + cha_mod
        else:
            print(save)
            
        
    return
    
```

```python
def gen_equipment(player_sheet):
    profs = player_sheet['Other Proficiencies & Languages']['Other proficiencies']
    weaps = player_sheet['Equipment']['Weapons & Armor']['Weapons']
    if player_sheet['Class'] == 'Barbarian':
        weaps.append(random.choice(['greataxe', 'martial melee weapon']))
        if 'greataxe' in weaps:
            if 'greataxe' in profs or 'Martial weapons' in profs:
                bonus = player_sheet['Proficiency bonus']
            else:
                bonus = 0
            player_sheet['Attacks & Spellcasting']['greataxe'] = {
                'Atk bonus': player_sheet['Ability Scores']['Strength'][1] + bonus,
                'Damage die': '1d12 slashing',
                'Dmg bonus': player_sheet['Ability Scores']['Strength'][1],
                'Properties': ['heavy', 'two-handed']
            }

        weaps.append(random.choice([{'handaxe': 2}, 'simple weapon']))
        weaps.append({'javelins': 4})
        player_sheet['Attacks & Spellcasting']['javelin'] = {
            'Atk bonus':player_sheet['Ability Scores']['Strength'][1] + player_sheet['Proficiency bonus'],
            'Damage die':'1d6 piercing',
            'Dmg bonus':player_sheet['Ability Scores']['Strength'][1],
            'Properties': [{'thrown': 'range 30/120'}]
        }
        player_sheet['Equipment']['Tools'].append("Explorer's pack")

    elif player_sheet['Class'] == 'Fighter':
        fstyle = None
        for itm in player_sheet['Features & Traits']:
            if type(itm) == dict:
                x = player_sheet['Features & Traits'].index(itm)
                if 'Fighting Style' in player_sheet['Features & Traits'][x]:
                    fstyle = player_sheet['Features & Traits'][x]
                    break

        if player_sheet['Ability Scores']['Dexterity'][1] + 11 > 16 or fstyle[
                'Fighting Style'] == 'Archery':
            player_sheet['Equipment']['Weapons & Armor']['Armor'].append(
                'Leather armor')
            weaps.append('Longbow')
            player_sheet['Attacks & Spellcasting']['longbow'] = {
            'Atk bonus':player_sheet['Ability Scores']['Dexterity'][1] + player_sheet['Proficiency bonus'],
            'Damage die':'1d8 piercing',
            'Dmg bonus':player_sheet['Ability Scores']['Dexterity'][1],
            'Properties': ['ranged', 'heavy', 'two-handed',{'ammunition': 'range 150/600'}]
        }
            weaps.append({'arrows': 20})
            player_sheet['Armor Class'] = player_sheet['Armor Class'] + \
                11 + player_sheet['Ability Scores']['Dexterity'][1]
        else:
            player_sheet['Equipment']['Weapons & Armor']['Armor'].append(
                'Chain mail')
            player_sheet['Armor Class'] = player_sheet['Armor Class'] + 16

        if fstyle['Fighting Style'] == 'Dueling' or fstyle[
                'Fighting Style'] == 'Protection':
            weaps.append('Martial weapon')
            player_sheet['Equipment']['Weapons & Armor']['Armor'].append('Shield')
            player_sheet['Armor Class'] = player_sheet['Armor Class'] + 2
        elif fstyle['Fighting Style'] == 'Two-Weapon Fighting':
            weaps.append('Martial weapon')
            weaps.append('Martial weapon')
        else:
            x = random.randint(0, 1)
            if x == 0:
                weaps.append('Martial weapon')
                player_sheet['Equipment']['Weapons & Armor']['Armor'].append(
                    'Shield')
                player_sheet['Armor Class'] = player_sheet['Armor Class'] + 2
            else:
                weaps.append('Martial weapon')
                weaps.append('Martial weapon')
        # should I have archery go for the light crossbow?
        weaps.append(random.choice(['light crossbow', {'handaxe': 2}]))
        for itm in weaps:
            if type(itm) == dict:
                if 'handaxe' in weaps[weaps.index(itm)]:
                    if 'handaxe' in profs or 'Simple weapons' in profs:
                        bonus = player_sheet['Proficiency bonus']
                    else:
                        bonus = 0
                    player_sheet['Attacks & Spellcasting']['handaxe'] = {
                        'Atk bonus':
                        player_sheet['Ability Scores']['Strength'][1] + bonus,
                        'Damage die':
                        '1d6 slashing',
                        'Dmg bonus':
                        player_sheet['Ability Scores']['Strength'][1],
                        'Properties': ['light', {
                            'thrown': 'range 20/60'
                        }]
                    }
                    break


        if 'light crossbow' in weaps:
            if 'light crossbow' in profs or 'Simple weapons' in profs:
                bonus = player_sheet['Proficiency bonus']
            else:
                bonus = 0

            player_sheet['Attacks & Spellcasting']['light crossbow'] = {
                        'Atk bonus':
                        player_sheet['Ability Scores']['Dexterity'][1] + bonus,
                        'Damage die':
                        '1d8 slashing',
                        'Dmg bonus':
                        player_sheet['Ability Scores']['Dexterity'][1],
                        'Properties': ['ranged', {'ammunition':'range 80/320'}, 'loading', 'two-handed']
                    }
    #         if fstyle == 'Archery':
    #             player_sheet['Attacks & Spellcasting']['light crossbow']['Atk bonus'] = player_sheet['Attacks & Spellcasting']['light crossbow']['Atk bonus'] + 2
            weaps.append({'bolts': 20})

        player_sheet['Equipment']['Tools'].append(
            random.choice(["Dungeoneer's pack", "Explorer's pack"]))

    return
```
